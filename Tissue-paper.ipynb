```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

x = np.array([[7, 7], [7, 4], [3, 4], [1, 4]])
y = np.array(['Bad', 'Bad', 'Good', 'Good'])

# Range of K values to test
k_params = {'n_neighbors': range(1, 11)}

# K-NN classifier
knn = KNeighborsClassifier()

# Perform GridSearchCV
grid_search = GridSearchCV(knn, k_params, cv=2)
grid_search.fit(x, y)

# Print the best k value and its score
best_k = grid_search.best_params_['n_neighbors']
score = grid_search.best_score_
mean_accuracy = np.mean(score)

#print(f"The best k value is {best_k} with a mean score of {score}")


#fit model
knn = KNeighborsClassifier(n_neighbors=best_k)
knn.fit(x,y)

new_tissue_paper = np.array([[3,7]])
predicted_quality = knn.predict(new_tissue_paper)

predicted_quality
```
